```
import time
def callTime(methodToRun, positional_arguments, verbose=True):
    start = time.time()
    result = methodToRun(positional_arguments)
    end = time.time()
    if verbose:
        print("Time is %ss" % (end-start))
    return result
```

```
import numpy as np
tamanho = 10000000

def vetorSeguidoComAppend(tamanho):
    vet = []
    for i in range(tamanho):
        vet.append(i + 1)

def vetorSeguidoComIndice(tamanho):
    vet = [0]*tamanho
    for i in range(tamanho):
        vet[i] = i + 1
    
def vetorSeguidoComNP(tamanho):
    vet = np.zeros(tamanho)
    for i in range(tamanho):
        vet[i] = i + 1

def vetorSeguidoLambda(tamanho):
    vet = [x + 1 for x in range(tamanho)]

def vetorSeguidoNPArange(tamanho):
    vet = np.arange(start=1, stop=tamanho+1)
    

callTime(vetorSeguidoComAppend, tamanho)
callTime(vetorSeguidoComIndice, tamanho)
callTime(vetorSeguidoComNP, tamanho)
callTime(vetorSeguidoLambda, tamanho)
callTime(vetorSeguidoNPArange, tamanho)
```

```
from numba import jit, njit, vectorize
#REFAZER
tamanho = 1000000
vetNp = np.arange(start=1, stop=tamanho+1)

@jit
def addingToVectorNumba(input_list1):
    value1 = 0.1
    tamanho1 = 1000000
    for i1, item1 in enumerate(input_list1):
        input_list1[i1] = item1 + value1/(tamanho1+1)
    return input_list1

@vectorize
def addingToVectorNumba2(num2):
    value2 = 0.1
    tamanho2 = 1000000
    return num2 + value2/(tamanho2 + 1)
    

def addingToVectorNumpyPro(input_list3):
    value3 = 0.1
    tamanho3 = 1000000
    def temp3(x):
        return value3/(tamanho3+1)
    vetAdd3 = np.fromfunction(temp3, [tamanho3])
    #vetAdd = vetAdd.reshape(tamanho, 1)
    input_list3 = np.add(input_list3, vetAdd3)
    return input_list3

vetNp1 = np.arange(start=1, stop=tamanho+1, dtype='float64')
callTime(addingToVectorNumba, vetNp1)
vetNp2 = np.arange(start=1, stop=tamanho+1, dtype='float64')
callTime(addingToVectorNumba2, vetNp2)
vetNp3 = np.arange(start=1, stop=tamanho+1, dtype='float64')
callTime(addingToVectorNumpyPro, vetNp3)
```
