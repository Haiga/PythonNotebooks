```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
import numpy as np


#y = 2x^2 + 3

tamanho_amostras = 50

X = np.arange(start=1, stop=tamanho_amostras+1)
variation = np.random.normal(0, 20, tamanho_amostras)
#def temp(indice):
#    return 2*(indice + 1)*(indice + 1) + 3

def temp(indice):
    return 2*(indice + 1) + 3

y_correto = np.fromfunction(temp, [tamanho_amostras])
y = np.add(y_correto, variation)
```

```
regr = LinearRegression()
regr.fit(X.reshape(-1, 1), y)
```

```
y_predict = regr.predict(X.reshape(-1, 1))
```

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(X, y, 'b.', label='Amostras')
ax.plot(X, y_correto, label='Função Real')
ax.plot(X, y_predict, 'k', label='Regressão predito')

leg = ax.legend()
```

```
KNN = KNeighborsRegressor(n_neighbors=3)
KNN.fit(X.reshape(-1, 1), y)

y_predict_knn = KNN.predict(X.reshape(-1, 1))
```

```
fig, ax = plt.subplots()

ax.plot(X, y, 'b.', label='Amostras')
ax.plot(X, y_correto, label='Função Real')
ax.plot(X, y_predict_knn, 'k', label='KNN predito n_neighbors=3')

leg = ax.legend()
```

```
import math
def rmse(y_real, y_predictted):
    n = np.size(y_real)
    return math.sqrt(np.sum((y_real - y_predictted)*(y_real - y_predictted) / n))
```

```
rmse(y, y_predict_knn)
```

```
rmse(y, y_predict)
```

```
rmse(y_correto, y_predict_knn)
```

```
rmse(y_correto, y_predict)
```

```
KNN = KNeighborsRegressor(n_neighbors=20)
KNN.fit(X.reshape(-1, 1), y)

y_predict_knn = KNN.predict(X.reshape(-1, 1))

fig, ax = plt.subplots()

ax.plot(X, y, 'b.', label='Amostras')
ax.plot(X, y_correto, label='Função Real')
ax.plot(X, y_predict_knn, 'k', label='KNN predito n_neighbors = 20')

leg = ax.legend()
```

```
rmse(y, y_predict_knn)
```
