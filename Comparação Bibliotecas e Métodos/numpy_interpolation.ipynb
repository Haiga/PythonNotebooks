```
#Utilizando interpolação linear, não é suave, como muitos pontos
#se aproxima do ideal, porém nem sempre temos muitos pontos.
import numpy as np
import matplotlib.pyplot as plt

xfine = np.linspace(-5,5,101)
yfine = np.sin(xfine*np.pi/2.5)

plt.close('all')

NN = [6,11,21];
Na = len(NN)

for k in range(Na):
    N = NN[k]
    x = np.linspace(-5,5,N)
    y = np.sin(x*np.pi/2.5)
    amostray = np.sin(x*np.pi/2.5)
    yinter = np.interp(xfine, x, y);
    plt.subplot(3,1,k+1)
    plt.plot(xfine, yfine, color='black')#original
    plt.scatter(x, y, color='blue') #amostra
    plt.plot(xfine, yinter, color='red') #interpolado
    plt.ylim(-1.5,1.5) 
```

```
#Utilizando interpolação polinomial a função ficará suave, porém com poucos pontos
#a tendência quadrática, cúbica ou linear da função causa expansão exagerada da função
import numpy as np
import matplotlib.pyplot as plt

xfine = np.linspace(-5,5,101)
yfine = np.sin(xfine*np.pi/2.5)

plt.close('all')

NN = [6,11,21];
Na = len(NN)

for k in range(Na):
    N = NN[k]
    x = np.linspace(-5,5,N)
    y = np.sin(x*np.pi/2.5)
    amostray = np.sin(x*np.pi/2.5)
    
    #p = np.polyfit(x, y, N-1)
    p = np.polyfit(x, y, N-1)
    yinter=np.polyval(p, xfine)

    plt.subplot(3,1,k+1)
    plt.plot(xfine, yfine, color='black')#original
    plt.scatter(x, y, color='blue') #amostra
    plt.plot(xfine, yinter, color='red') #interpolado
    plt.ylim(-1.5,1.5) 
```

```
#Com splines função fica suave, e são exploradas tendências cúbicas e quadráticas.
#São considerados os pontos mais próximos, mantendo uma tendência linear em próximos pontos
#Utilizadas derivadas de várias ordens
#See docs: https://docs.scipy.org/doc/scipy/reference/interpolate.html#module-scipy.interpolate
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import *

xfine = np.linspace(-5,5,101)
yfine = np.sin(xfine*np.pi/2.5)

plt.close('all')

NN = [6,11,21];
Na = len(NN)

for k in range(Na):
    N = NN[k]
    x = np.linspace(-5,5,N)
    y = np.sin(x*np.pi/2.5)
    amostray = np.sin(x*np.pi/2.5)
    
    yinter = interp1d(x, y)

    plt.subplot(3,1,k+1)
    plt.plot(xfine, yfine, color='black')#original
    plt.scatter(x, y, color='blue') #amostra
    plt.plot(x, yinter.y, color='red') #interpolado
    plt.ylim(-1.5,1.5) 
```
