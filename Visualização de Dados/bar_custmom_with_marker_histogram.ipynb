```
import matplotlib.pyplot as plt
import numpy as np

fig = plt.subplots()


total_bars = 20

x = np.arange(total_bars)
y = np.random.uniform(0, 3, total_bars)

y_down = np.random.uniform(-3, 0, total_bars)


def my_bar_with_many_markers(positionx, valuey, interval=0.1, c='b'):
    if valuey > 0:
        balls = np.arange(0, valuey, interval)
    else:
        balls = np.arange(valuey, 0, interval)
    positions = np.array([positionx]*balls.size)
    size_balls = [10]*balls.size
    plt.scatter(positions, balls, size_balls, c=c)
    plt.scatter(positions+0.2, balls, size_balls, c=c)
    plt.scatter(positions-0.2, balls, size_balls, c=c)
    # plt.scatter(positions-0.2, balls, size_balls, c='b', marker='*')
    # plt.scatter(positions-0.2, balls, size_balls, c='b', marker='$oi$')

def my_bar_with_one_marker(positionx, valuey, interval=0.12, c='b'):
    if valuey > 0:
        balls = np.arange(0, valuey, interval)
    else:
        balls = np.arange(valuey, 0, interval)
    positions = np.array([positionx]*balls.size)
    size_balls = [10]*balls.size
    plt.scatter(positions, balls, size_balls, c=c)


for xx, yy in zip(x, y):
    # plt.bar(xx, yy, color='b')
    my_bar_with_one_marker(xx, yy, c='#000000')


for xx, yy in zip(x, y_down):
    my_bar_with_one_marker(xx, yy, c='k')
    # plt.bar(xx, yy, color='k')

plt.xlim(-1, total_bars)
minimo = np.min(y_down) - 0.1
maximo = np.max(y) + 0.1
plt.ylim([minimo, maximo])

plt.fill_between([-1, total_bars/3], [maximo, maximo], color='b', alpha=0.2)
plt.fill_between([-1, total_bars/3], [minimo, minimo], color='b', alpha=0.2)

plt.fill_between([total_bars/3, 2*total_bars/3], [maximo, maximo], color='g', alpha=0.2)
plt.fill_between([total_bars/3, 2*total_bars/3], [minimo, minimo], color='g', alpha=0.2)

plt.fill_between([2*total_bars/3, total_bars], [maximo, maximo], color='r', alpha=0.2)
plt.fill_between([2*total_bars/3, total_bars], [minimo, minimo], color='r', alpha=0.2)

plt.xticks([])
plt.show()
```
