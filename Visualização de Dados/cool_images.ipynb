```
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))

angs = np.linspace(0, 2*np.pi, 50)

xc = 5
yc = 5
r = 2

x = r*np.cos(angs) + xc
topCircle = r*np.sin(angs) + yc
plt.plot(x, topCircle, 'k--', linewidth=1)

for i in range(9):
    xc = x[i*5]
    yc = topCircle[i*5]
    x = r*np.cos(angs) + xc

    topCircle = r*np.sin(angs) + yc
    plt.plot(x, topCircle, 'k--', linewidth=1)

# plt.xlim([0, 10])
# plt.ylim([0, 10])
```
