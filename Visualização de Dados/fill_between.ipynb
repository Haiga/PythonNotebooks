```
import matplotlib.pyplot as plt
import numpy as np

fig, ax1 = plt.subplots()


x = np.arange(75)

y_mean = np.random.uniform(0, 1, 75)
y_above = y_mean + np.random.uniform(0, 4, 75)
y_below = y_mean - np.random.uniform(0, 4, 75)

ax1.plot(x, y_mean, ls='--', label='Mean')

ax1.fill_between(x, y_mean, y_above, alpha=0.2)
ax1.fill_between(x, y_below, y_mean, alpha=0.2)


ax1.legend(ncol=3, loc='lower right')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

plt.show()
```
