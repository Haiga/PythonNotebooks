```
import matplotlib.pyplot as plt

img = plt.imread("img.jpg")
fig, ax = plt.subplots()
x = range(300)
#É possível definir, tamanho e posicação com extent
#[x_inicial, xfinal, y_inicial, y_final]
ax.imshow(img, extent=[100, 200, 300, 400])
ax.imshow(img, extent=[200, 300, 300, 400])
ax.imshow(img, extent=[200, 300, 200, 300])
ax.imshow(img, extent=[0, 100, 300, 400])
ax.imshow(img, extent=[100, 200, 300, 400])

from matplotlib import transforms
rotation_in_degrees = -45
#É possível transformar a imagem com o matplotlib transforms 
#(embora talvez haja necessidade de aplicar ideias de Computação Gráfica para que fique conforme desejado)
tr = transforms.Affine2D().rotate_deg(rotation_in_degrees)
ax.imshow(img, transform=tr + ax.transData, extent=[100, 200, 200, 300])
#ax.set_xlim([-300, 300])
ax.plot(x, x, '--', linewidth=5, color='firebrick')
```
