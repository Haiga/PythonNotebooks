```
import matplotlib.pylab as plt
markers=['.',',','o','v','^','<','>','1','2','3','4','8','s','p','P','*','h','H','+','x','X','D','d','|','_']
descriptions=['point', 'pixel', 'circle', 'triangle_down', 'triangle_up','triangle_left', 'triangle_right', 'tri_down', 'tri_up', 'tri_left','tri_right', 'octagon', 'square', 'pentagon', 'plus (filled)','star', 'hexagon1', 'hexagon2', 'plus', 'x', 'x (filled)','diamond', 'thin_diamond', 'vline', 'hline']
x=[]
y=[]
for i in range(5):
    for j in range(5):
        x.append(i)
        y.append(j)
plt.figure()
for i,j,m,l in zip(x,y,markers,descriptions):
    plt.scatter(i,j,marker=m)
    plt.text(i-0.15,j+0.15,s=m+' : '+l)
plt.axis([-0.1,4.8,-0.1,4.5])
plt.tight_layout()
plt.axis('off')
plt.show()  
```

```
def getMarker(description):
    markers=['.',',','o','v','^','<','>','1','2','3','4','8','s','p','P','*','h','H','+','x','X','D','d','|','_']
    descriptions=['point', 'pixel', 'circle', 'triangle_down', 'triangle_up','triangle_left', 'triangle_right', 'tri_down', 'tri_up', 'tri_left','tri_right', 'octagon', 'square', 'pentagon', 'plus (filled)','star', 'hexagon1', 'hexagon2', 'plus', 'x', 'x (filled)','diamond', 'thin_diamond', 'vline', 'hline']

    return markers[descriptions.index(description)]
```

```
getMarker('point')
```

```
#Example usage 
#Credits to pydatascience: https://pydatascience.org/2017/11/24/plot-multiple-lines-in-one-chart-with-different-style-python-matplotlib/
import matplotlib.pyplot as plt
import numpy as np
 
x=np.arange(6)
 
fig=plt.figure()
ax=fig.add_subplot(111)
 
ax.plot(x,x,c='b',marker="^",ls='--',label='Greedy',fillstyle='none')
ax.plot(x,x+1,c='g',marker=(8,2,0),ls='--',label='Greedy Heuristic')
ax.plot(x,(x+1)**2,c='k',ls='-',label='Random')
ax.plot(x,(x-1)**2,c='r',marker="v",ls='-',label='GMC')
ax.plot(x,x**2-1,c='m',marker="o",ls='--',label='KSTW',fillstyle='none')
ax.plot(x,x-1,c='k',marker="+",ls=':',label='DGYC')
 
plt.legend(loc=2)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(50, step=0.1)
y = np.sin(x)
ycos = np.cos(x)

#Quando se tem muitos pontos os markers começam a atrapalhar se não usados corretamente
plt.plot(x, y, marker=getMarker('plus'))
plt.plot(x, ycos, marker=getMarker('triangle_down'))
```

```
#Para utilizar Markers e ainda manter uma boa visualização basta simplesmete 
#colocá-los mais espaçados

import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(50, step=0.1)
y = np.sin(x)
ycos = np.cos(x)

#O vetor x tem 500 elementos, vamos colocar na função seno slices a cada 20 elementos
points_interval = 20
my_slice = points_interval*np.arange(int(500/points_interval))
plt.plot(x, y)
plt.scatter(x[my_slice], y[my_slice], marker=getMarker('plus'))

#O vetor x tem 500 elementos, vamos colocar na função cosseno slices a cada 12 elementos
points_interval = 14
my_slice = points_interval*np.arange(int(500/points_interval))

plt.plot(x, ycos)
plt.scatter(x[my_slice], ycos[my_slice], marker=getMarker('plus'))
```
