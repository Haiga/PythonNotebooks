```
import plotly.express as px
import pandas as pd

#Comparar um mesmo algoritmo em configurações diferentes, exemplo métricas diferentes
df = pd.DataFrame(
    dict(
    r=[0.1, 0.5, 0.2, 0.32],
    theta=['Nov.Div.Efet.','Nov.Efet.','Nov.Div.',
           'Div.Efet.']))
#Esse fecha a figura, mas não procurei implementação desse método com múltiplos outros 
#polígonos, pode ser que tenha
fig = px.line_polar(df, r='r', theta='theta', line_close=True)
fig.update_traces(fill='toself')
fig.show()
```

```
#Documentação:
#https://plot.ly/python-api-reference/generated/plotly.graph_objects.Scatter.html


import plotly.graph_objects as go

#Adicionei a mesma categoria do início no final, pois as linhas do polígono não 
#estavam conectando, #Altém disso, para funcionar adicionei em todos o r (arrays)
#o mesmo valor da posição inicial no final
categories = ['Nov.Div.Efet.','Nov.Efet.','Nov.Div.',
           'Div.Efet.', 'Nov.Div.Efet.']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[0.1, 0.5, 0.2, 0.32, 0.1],
      theta=categories,
      fill='toself',
      name='Web10k',
      marker = dict(
        color = "#8090c7",
        symbol = "square",
        size = 8
      ),
      
))
fig.add_trace(go.Scatterpolar(
      r=[0.4, 0.3, 0.5, 0.1, 0.4],
      theta=categories,
      fill='toself',#É usado para preencher o interior do polígono
      name='Movielens',
      #connectgaps=True
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
        visible=True,# To show 0, 0.1, 0.2 ... e as grades
        #range=[0, 5] # seta o raio do gráfico,
        
    )),
  showlegend=True,
)

fig.show()
```
