```
#Vamos simular um experimento variando em 100s
#Valores para representar probabilidade
from random import *
import numpy as np
import matplotlib.pyplot as plt


time = 100
y1 = [1/(100-x) for x in range(time)]
y2 = [(x*x)/(100*100) for x in range(time)]
y3 = [x/100 + 5*uniform(0,0.01)*uniform(0,0.01) for x in range(time)]
y4 = [1/(100-x) -x*0.001 for x in range(time)]
x = [(x+1) for x in range(time)]
 
plt.plot( x, y1[::-1], 'g') # green bolinha
plt.plot( x, y2[::-1], 'k:', color='orange') # linha pontilha orange
plt.plot( x, y3[::-1], 'r.') # red triangulo
plt.plot( x, y4[::-1], 'k--', color='blue')  # linha tracejada azul

#plt.axis([-10, 60, 0, 11])
plt.title("Mais incrementado")

plt.xlabel("eixo horizontal")
plt.ylabel("eixo vertical")
#plt.margins(x=1, y=1)
plt.show()
```
