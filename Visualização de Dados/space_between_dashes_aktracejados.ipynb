```
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3]
y = np.array([0,4, 9])

plt.plot(x, y * 0,       c='#000000', label='1')
plt.plot(x, y * 1, '--', c='#3E424B', dashes=(5, 1), label='2')
plt.plot(x, y * 2, '--', c='#696969', dashes=(10, 1), label='3')
plt.plot(x, y * 3, '--', c='#808080', dashes=(5, 5), label='4')
plt.plot(x, y * 4, '--', c='#A9A9A9', dashes=(2, 2), label='5')
plt.plot(x, y * 5, '--', c='#363636', dashes=(1, 1), label='6')

plt.legend()
```
