```
import matplotlib.pyplot as plt
import numpy as np


begin = np.array([2003,1991,2008,1986,2013,1994,2002,2030])
end =   np.array([2007,2016,2016,2015,2014,1999,2008,2050])

event = ["Event {}".format(i) for i in range(len(begin))]

#Plota várias barras horizontais
#O primeiro parâmetro é a posição no eixo y de cada uma das barras 
#(no plot na próxima célula é póssivel ver como mudar posição dos eventos)
#O segundo parâmetro é um array com o tamanho de cada barra (end - begin retorna um array pois é um np.array)
#Os demais parâmetros são opicionais, porém para fazer uma "timeline"
#não basta informar o tamanho das barras, pois todas, começariam no valor padrão (0)
#Pode ser informado o início das barras com o parâmetro left (que é um array contendo os respectivos começos das barras)
plt.barh(range(len(begin)),  end-begin, left=begin)

plt.xlim((2000,2050))

#Uma forma fácil de definir os labels (yticks ou xticks) é passar suas
#posições (primeiro parâmetro) e respectivos valores (segundo parâmetro)
plt.yticks(range(len(begin)), event)
plt.show()
```
